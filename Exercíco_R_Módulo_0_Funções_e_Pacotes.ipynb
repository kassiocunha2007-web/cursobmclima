### Exercício 1

Imagine que você está realizando a gestão de três programas de saúde em uma secretaria estadual. Utilizando o RStudio, escreva um script que realize as seguintes tarefas:

* Dados de entrada (variáveis):

  * Nome dos programas
  * Objetivos dos programas
  * Público-alvo dos programas
  * Orçamento anual de cada programa
  * Datas de início e fim de cada programa

* Tarefas:

  * Criar uma lista com as informações de cada programa
  * Calcular o orçamento total destinado a todos os programas
  * Identificar o programa com o maior orçamento
  * Listar os programas que estão em andamento atualmente

* Desafio:
  * Exibir o programa com a data de início mais recente

Orientações:

- O primeiro objetivo desta atividade é treinar a atribuição de tipos de dados às variáveis e contantes. Então, declare uma variável separada para cada dado de entrada solicitada, utilizando um tipo de dados adequado ao que deseja atribuir ou representar. Exemplo, se a variável é orçamento o tipo de dados deverá ser numérico.
- O segundo objetivo é exercitar o uso de operadores. Então, busque testar os operadores que melhor se adequam às tarefas de processamento solicitadas. Por exemplo, a quantidade total de orçamento normalmente é a soma do valor de todos os programas.
- Procure fazer os exercícios antes de procurar por soluções prontas. Lembre-se que não estamos realizando uma "prova" mas um treinamento. Aqui o processo de aprendizado é mais importante que o resultado pronto.

```r
nomes <- c("Programa A", "Programa B", "Programa C")

objetivos <- c(
  "Reduzir doenças crônicas",
  "Ampliar cobertura vacinal",
  "Melhorar saúde materno-infantil"
)
```

```r
publico_alvo <- c(
  "Adultos",
  "Crianças",
  "Gestantes"
)
```

```r
orcamento <- c(1000000, 1500000, 1200000)

data_inicio <- as.Date(c("2023-01-01", "2024-03-01", "2025-01-15"))
data_fim    <- as.Date(c("2026-12-31", "2025-12-31", "2027-06-30"))
```

* *Lista com as informações de cada programa*

```r
programas <- list(
  nomes = nomes,
  objetivos = objetivos,
  publico_alvo = publico_alvo,
  orcamento = orcamento,
  data_inicio = data_inicio,
  data_fim = data_fim
)
```

```r
programas
```

* *Orçamento total*

```r
sum(orcamento)
```

* *Programa com o maior orçamento*

```r
nomes[which.max(orcamento)]
```

* *Programas que estão em andamento atualmente*

```r
hoje <- Sys.Date()

nomes[data_inicio <= hoje & data_fim >= hoje]
```

* *Desafio: programa com a data de início mais recente*

```r
nomes[which.max(data_inicio)]
```
